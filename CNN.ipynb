```python
import tensorflow as tf
from keras import layers,models,regularizers,datasets
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow import keras
from keras.layers import Flatten,Dense,Conv2D,MaxPooling2D,Dropout
from keras.models import Sequential
from keras.datasets import cifar10
import cv2
```

```python
(X_train,y_train),(X_test,y_test)=datasets.cifar10.load_data()
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python
X_test=X_test/255
X_train=X_train/255
```

```python
from keras.src.layers import BatchNormalization
model=Sequential([



layers.Conv2D(32,kernel_size=(3,3),padding="valid",kernel_regularizer=regularizers.l2(0.001),input_shape=(32,32,3)),
layers.BatchNormalization(),
layers.ReLU(),
layers.Dropout(0.3),

layers.Conv2D(32,kernel_size=(3,3),padding="valid"),
layers.BatchNormalization(),
layers.ReLU(),
layers.MaxPooling2D(pool_size=(2,2)),

layers.Conv2D(64, (3, 3), padding='valid',
                  kernel_regularizer=regularizers.l2(0.001)),
layers.BatchNormalization(),
layers.ReLU(),
layers.Dropout(0.4),

layers.Conv2D(64, (3, 3), padding='valid',
                  kernel_regularizer=regularizers.l2(0.001)),
layers.BatchNormalization(),
layers.ReLU(),
layers.MaxPooling2D((2, 2)),

layers.Flatten(),

layers.Dense(128,kernel_regularizer=regularizers.l2(0.001)),
layers.BatchNormalization(),
layers.ReLU(),
layers.Dropout(0.5),

layers.Dense(10,activation='softmax')

])
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(X_train,y_train,epochs=10,validation_data=(X_test,y_test),verbose=1)
```

```python
loss,accuracy=model.evaluate(X_test,y_test)
```
